```python
#import all dependenices

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
#Scikit learn python's standard ML library(pre-installed in colab)

from sklearn.model_selection import train_test_split
#train_test_split divides the data into training and testing data

from sklearn.preprocessing import LabelEncoder,StandardScaler
#Lable Encode is uses to change the catogorical data into numerical value . It starts with (0)
#StandardScalar: scalar numbers to all features have same range

from sklearn.linear_model import LinearRegression
#Linearregression :fits a straight line relationship between features and target

from sklearn.tree import DecisionTreeRegressor
#decision tree regressors : build an tree if/else rule to predict

from sklearn.ensemble import RandomForestRegressor
#random forest regressors : builds 100 decision trees and give an average prediction

from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
# mean_absolute_error : avaerage absolute error for prediction
#mean_squared_error: average squaerd error (penalities for large mistake)
# r2_score :0.0 to 1.0 quality score (1.0 is perfect)

print("All libraries are imported sucessfully")
```

```python
#load the student dataset

df = pd.read_csv('/content/drive/MyDrive/Copy of student_performance.csv')
print(f'Dataset loaded:{df.shape[0]}students,{df.shape[1]}')
print(f'Colums list{df.columns.tolist()}')
df.head(5)
```

```python
#step 3 missing values

print(f'Missing values:\n{df.isnull().sum()}')
print(f'Total Missing values in the whole dataset {df.isnull().sum().sum()}')
```

```python
df_ml=df.copy()
le_gender=LabelEncoder()
df_ml['gender_encoder']=le_gender.fit_transform(df_ml['gender'])
print(f'Gender Encoding:{dict(zip(le_gender.classes_,le_gender.fit_transform(le_gender.classes_)))}')
le_dept = LabelEncoder()
df_ml['dept_encoder']=le_dept.fit_transform(df_ml['department'])
print(f'Department Encoding:{dict(zip(le_dept.classes_,le_dept.transform(le_dept.classes_)))}')
print('\nNew Columns added in the dataset:dept_encoder,gender_encode')
df_ml[['gender','gender_encoder','department','dept_encoder']].head(5)
```

```python
#feature selection

features_col =[
    'math_score',
    'science_score',
    'english_score',
    'attendance_percentage',
    'gender_encoder',
    'dept_encoder',
]
X=df_ml[features_col]
Y=df_ml['programming_score']

print(f'Feature Matrix X shape:{X.shape}(Students*features)')
print(f'Target Vector Y shape:{Y.shape}(one score per student)')
print(f'\nFeatures columns{features_col}')
print(f'Target Column:programming score')
print(f'\n Target range: {Y.min()} to {Y.max()} (mean:{Y.mean():.1f})')
```

```python
# step 6 -Train/Test split

X_train,X_test,Y_train,Y_test=train_test_split(
    X,Y,
  test_size=0.2,
  random_state=42)

print(f'Total Students:{len(X)}')
print(f'Traning Students:{len(X_train)} ({len(X_train)/len(X) *100:.0f}%)')
print(f'Testing Students:{len(X_test)} ({len(X_test)/len(X) *100:.0f}%)')
print(f'\nTraning target range:{len(Y_train)} ({len(Y_train)/len(Y) *100:.0f}%)')
print(f'Testing target range:{len(Y_test)} ({len(Y_test)/len(Y) *100:.0f}%)')
```

```python
le_standard = StandardScaler()
X_train_scaled = le_standard.fit_transform(X_train)
X_test_scaled = le_standard.transform(X_test)

print(f'Features scaled succesfully !')
print(f'Training feature mean(should be =0):{X_train_scaled.mean():.4f}')
print(f'Training feature std(should be =1):{X_train_scaled.std():.4f}')
print('\nNote:Scaler was fit on trainig data only-prevent data leakage')
```

```python
#step 9 : Train model 2 : Decision Tree

dt_model = DecisionTreeRegressor(
    max_depth=5,
    random_state=42
)
dt_model.fit(X_train_scaled, y_train)
dt_pred = dt_model.predict(X_test_scaled)
dt_mean = mean_absolute_error(y_test, dt_pred)
dt_rmse = np.sqrt(mean_squared_error(y_test, dt_pred))
dt_r2 = r2_score(y_test, dt_pred)

print('=== Model 2: Decision Tree ===')
print(f'MAE: {dt_mean:.2f} (prediction are off by {dt_mean:.1f} points on average)')
print(f'RMSE: {dt_rmse:.2f} (error with penalty for large mistakes)')
print(f'R^2: {dt_r2:.4f} ({dt_r2*100:.1f}% of programming score variation explained)')
```

```python
#step 9 : Train model 3 : Random Forest

dt_model = RandomForestRegressor(
    max_depth=5,
    random_state=42
)
dt_model.fit(X_train_scaled, y_train)
dt_pred = dt_model.predict(X_test_scaled)
dt_mean = mean_absolute_error(y_test, dt_pred)
dt_rmse = np.sqrt(mean_squared_error(y_test, dt_pred))
dt_r2 = r2_score(y_test, dt_pred)

print('=== Model 3 :Random forest ===')
print(f'MAE: {dt_mean:.2f} (prediction are off by {dt_mean:.1f} points on average)')
print(f'RMSE: {dt_rmse:.2f} (error with penalty for large mistakes)')
print(f'R^2: {dt_r2:.4f} ({dt_r2*100:.1f}% of programming score variation explained)')
```
